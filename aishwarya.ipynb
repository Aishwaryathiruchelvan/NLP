```python
def create_unigram_corpus(text):
  tokens = word_tokenize(text.lower())
  unigrams = Counter(tokens)
  return unigrams
text = """AI is the ability of a machine to display human-like capabilities such as reasoning, learning, planning and creativity. AI enables technical systems to perceive their environment, deal with what they perceive, solve problems and act to achieve a specific goal."""
unigram_corpus = create_unigram_corpus(text)

print(unigram_corpus)
```

```python
import nltk
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.util import bigrams
nltk.download('punkt')
nltk.download('stopwords')
```

```python
text_corpus = """AI is the ability of a machine to display human-like capabilities such as reasoning, learning, planning and creativity. AI enables technical systems to perceive their environment, deal with what they perceive, solve problems and act to achieve a specific goal."""
tokens = word_tokenize(text_corpus)
tokens = [token.lower() for token in tokens]
stop_words = set(stopwords.words('english'))
tokens = [token for token in tokens if token.isalpha() and token not in stop_words]
bigrams_list = list(bigrams(tokens))
bigram_counts = Counter(bigrams_list)
print(bigram_counts)
```

```python
import nltk
from nltk import ngrams
from collections import Counter
nltk.download('punkt')
text = """AI is the ability of a machine to display human-like capabilities such as reasoning, learning, planning and creativity. AI enables technical systems to perceive their environment, deal with what they perceive, solve problems and act to achieve a specific goal."""
tokens = nltk.word_tokenize(text)
trigrams = list(ngrams(tokens, 3))
trigram_freq = Counter(trigrams)
for trigram, freq in trigram_freq.items():
  print(f"{trigram}: {freq}")
```

```python
import nltk
from nltk import ngrams
from collections import Counter
nltk.download('punkt')
```

```python
text = """AI is the ability of a machine to display human-like capabilities such as reasoning, learning, planning and creativity.
          AI enables technical systems to perceive their environment, deal with what they perceive, solve problems and act to achieve a specific goal."""
tokens = nltk.word_tokenize(text)
bigrams = list(ngrams(tokens, 2))
bigram_freq = Counter(bigrams)
word_freq = Counter(tokens)
bigram_probabilities = {}
for (w1, w2), count in bigram_freq.items():
  prob = count / word_freq[w1]
  bigram_probabilities[(w1, w2)] = prob
for bigram, prob in bigram_probabilities.items():
  print(f"P({bigram[1]} | {bigram[0]}) = {prob:.4f}")
```

```python
import nltk
from nltk.util import ngrams
from collections import defaultdict
import random
nltk.download('punkt')
```

```python
corpus = [
"I love programming in Python",
"Python is great for data science",
"I enjoy learning new programming languages",
"Data science is an exciting field",
"Programming is fun and rewarding"
]
tokenized_corpus = [nltk.word_tokenize(sentence.lower()) for sentence in corpus]
n = 2
n_grams = [ngrams(sentence, n) for sentence in tokenized_corpus]
```

```python
next_word_dict = defaultdict(list)
for sentence in n_grams:
  for n_gram in sentence:
    prefix, next_word = n_gram[0], n_gram[1]
    next_word_dict[prefix].append(next_word)
def predict_next_word(prefix):
  prefix = prefix.lower()
  if prefix in next_word_dict:
    return random.choice(next_word_dict[prefix])
  else:
    return "No prediction available."
input_prefix = "I love"
predicted_word = predict_next_word(input_prefix)
print(f"Next word prediction for '{input_prefix}': {predicted_word}")
```
